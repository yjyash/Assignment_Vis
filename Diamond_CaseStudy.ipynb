```python
# data['color'].values_count()
data.drop(['depth'],axis=1,inplace=True)
#we remove depth attribute as it is not related to any other value
```

```python
data.drop(['table'],axis=1,inplace=True)

#We remove table attribute as it is not related to any other value
```

```python
data.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(),annot=True)

#Therefore seen from heatmap we find price, carat, x, y, z heavily depend on each other at an average correlation value of 0.95

#Also x depends on y and x also depends on z

#And x increases then carat increases
```

```python
sns.pairplot(data)
```

```python
sns.relplot(x=data['price'],y=data['carat'],hue=data['x'],kind='scatter')

#Therefore price increases with increase in carat and x value.
```

```python
sns.relplot(x=data['x'],y=data['y'],kind='line')

#This shows x value increase with increase in y. Hence price is directly proportional to y.
```

```python
sns.relplot(x=data['x'],y=data['z'],kind='scatter')
```

```python
sns.relplot(x=data['y'],y=data['z'],kind='line')

#Therefore x and y are not much  related to eachother.
```

```python
sns.barplot(x=data['cut'],y=data['price'])

#Therefore it can be seen quality of cut increases the price of the diamond.

#Also ideal category of diamond cut is cheap among all.
```

```python
sns.barplot(x=data['color'],y=data['price'])

#Therefore we can see J is the best category for diamond and shows it is expensive.

#Also E is cheaper due to its poor color qualities.
```

```python
sns.barplot(x=data['clarity'],y=data['price'])

#As it can be seen from barplot that SI2 is the best clarity and expensive among all

#The worst clarity is offerd by WS1 category as it is cheaper among all.
```
